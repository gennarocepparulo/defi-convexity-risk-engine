# DeFi Convexity Risk - Exploration Notebook

This notebook explores LP positions and convexity risk analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from src.lp_model import value_lp, delta_gamma_lp
from src.price_fetcher import get_coingecko_price

# Fetch real price
S0 = get_coingecko_price('ethereum')
print(f'Current ETH price: ${S0}')
```

```python
# Define LP parameters
P_a, P_b = 1800, 2200  # Price range
L = 1000  # Liquidity

# Create price range for analysis
prices = np.linspace(1500, 2500, 100)
values = [value_lp(P, P_a, P_b, L) for P in prices]
deltas = [delta_gamma_lp(P, P_a, P_b, L)[0] for P in prices]
gammas = [delta_gamma_lp(P, P_a, P_b, L)[1] for P in prices]
```

```python
# Plot LP Value
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(prices, values, 'b-', linewidth=2)
plt.axvline(S0, color='r', linestyle='--', label=f'Current: ${S0:.0f}')
plt.xlabel('Price (USD)')
plt.ylabel('LP Value')
plt.title('LP Position Value')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 2)
plt.plot(prices, deltas, 'g-', linewidth=2)
plt.axvline(S0, color='r', linestyle='--')
plt.xlabel('Price (USD)')
plt.ylabel('Delta')
plt.title('Delta (Price Sensitivity)')
plt.grid(True, alpha=0.3)

plt.subplot(1, 3, 3)
plt.plot(prices, gammas, 'm-', linewidth=2)
plt.axvline(S0, color='r', linestyle='--')
plt.xlabel('Price (USD)')
plt.ylabel('Gamma')
plt.title('Gamma (Delta Sensitivity)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
